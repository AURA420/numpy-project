Task 1: Load and Inspect the Dataset
1. Load the dataset into a numpy array.
2. Inspect the first few rows of the dataset to understand its structure.
3. Check for any missing values and handle them appropriately.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Path to your data CSV file
file_path = 'data.csv'

# Read the CSV file into a NumPy array
data = np.genfromtxt(file_path, delimiter=',', skip_header=1)

# Display the array
print(data)
```

```python
# Fill missing values with 0
data_filled_zeros = np.where(np.isnan(data), 0, data)
```

```python
data_filled_zeros
```

Task 2: Basic Statistical Analysis
1. Calculate the mean, median, and standard deviation for each column in the dataset.
2. Identify the column with the highest mean value.

```python
d_mean=np.mean(data_filled_zeros,axis=0)
d_mean
```

```python
d_median=np.median(data_filled_zeros,axis=0)
d_median
```

```python
d_std=np.std(data_filled_zeros,axis=0)
d_std
```

```python
# get the column no where the mean is max
d=np.argmax(d_mean,axis=0)
d
# get the data of that column
data_filled_zeros[:,d]
```

Task 3: Data Manipulation
1. Normalize the data in the dataset so that each column has a mean of 0 and a standard deviation of 1.
2. Reshape the dataset into a 2D array where each row represents an observation and each column represents a feature.

```python
# Normalize the dataset
data_normalized = (data - d_mean) / d_std
data_normalized
```

```python

# Handle any potential NaN values that might result from normalization (e.g., dividing by zero std)
data_normaliz = np.nan_to_num(data_normalized)
data_normaliz
```

```python
data_normaliz.shape
```

Task 4: Advanced Analysis
1. Compute the correlation matrix for the dataset.
2. Identify any pairs of columns that have a correlation coefficient greater than 0.8 or less than -0.8.

```python
# Compute the correlation matrix
correlation_matrix = np.corrcoef(data_normaliz, rowvar=False)
correlation_matrix
```

```python
# Identify pairs of columns with correlation coefficient greater than 0.8 or less than -0.8
highly_correlated_pairs = []

num_columns = correlation_matrix.shape[0]

for i in range(num_columns):
    for j in range(i + 1, num_columns):
        if correlation_matrix[i, j] > 0.8 or correlation_matrix[i, j] < -0.8:
            highly_correlated_pairs.append((i, j, correlation_matrix[i, j]))
```

```python
# Print highly correlated pairs
for pair in highly_correlated_pairs:
    print(f"Columns {pair[0]} and {pair[1]} with correlation coefficient {pair[2]:.2f}")
```

Task 5: Data Visualization (Optional)
1. Use matplotlib or any other visualization library to plot the distribution of values for each column.
2. Create a heatmap of the correlation matrix.

```python
# Plot the distribution of values for each column
num_columns = data_normalized.shape[1]
fig, axes = plt.subplots(nrows=num_columns, ncols=1, figsize=(10, num_columns * 3))

for i in range(num_columns):
    axes[i].hist(data_normalized[:, i], bins=20, edgecolor='black')
    axes[i].set_title(f'Distribution of Column {i}')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            xticklabels=[f'Column {i}' for i in range(num_columns)], 
            yticklabels=[f'Column {i}' for i in range(num_columns)])
plt.title('Correlation Matrix Heatmap')
plt.show()
```
